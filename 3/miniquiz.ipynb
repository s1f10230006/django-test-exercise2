```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist = {0: 0.125, 1: 0.375, 2: 0.375, 3: 0.125}
```

```python
plt.bar(dist.keys(), dist.values(), width=1.0, ec='black')
plt.xlabel("Random variable X (Number of Heads)")
plt.ylabel("Probability")
plt.title("Probability Distribution")
plt.show()
```

```python
def Cdf(dist, x):
    prob = 0.0
    for value in dist.keys():
        if value <= x:
            prob += dist[value]
    return prob
```

```python
cdf_dis = {}
for x in range(0, 4):
    cdf_dis[x] = Cdf(dist, x)
```

```python
plt.bar(cdf_dis.keys(), cdf_dis.values(), width=1.0, ec='black')
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Cumulative Distribution Function")
plt.show()
```

```python
def Expect(dist):
    e = sum([x * p for x, p in dist.items()])
    return e
```

```python
def Variance(dist):
    e = Expect(dist)
    var = sum([(x - e)**2 * p for x, p in dist.items()])
    return var
```

```python
def Std(dist):
    return np.sqrt(Variance(dist))
```

```python
print("期待値: {:.2f}".format(Expect(dist)))
print("分散: {:.2f}".format(Variance(dist)))
print("標準偏差: {:.2f}".format(Std(dist)))
```
