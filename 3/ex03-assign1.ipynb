### 1. パッケージをインポートする Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 確率変数と確率分布をファイルから読み込む Load the R.V. and probability 

```python
df = pd.read_csv('data/rand_distribution.csv', skiprows=1, header=None)
data_dict = dict(df.to_numpy())

x = np.array(list(data_dict.keys()), dtype=float)
p = np.array(list(data_dict.values()), dtype=float)
```

### 辞書型に変換　Convert to dictionary type

```python
data_dict = dict(df.to_numpy())
x = np.array(list(data_dict.keys()), dtype=float)
p = np.array(list(data_dict.values()), dtype=float)
```

### 2. 棒グラフを使って確率分布を図示しなさい Plot the probability distribution by using a bar graph

```python
plt.bar(x, p)
plt.show()
```

### 3. 累積分布を定義しなさい Define a function of the CDF

```python
def get_cdf(p):
    return np.cumsum(p)
```

###  累積分布を求めなさい Calculate the CDF

```python
p_cum = get_cdf(p)
```

###  累積分布を図示しなさい Plot the CDF

```python
plt.step(x, p_cum, where='mid')
plt.show()
```

### 4. 期待値を求める関数を定義しなさい Define a function of expected value

```python
def get_expectation(x, p):
    return np.sum(x * p)
```

###  分散を求める関数を定義しなさい Define a function of variance

```python
def get_variance(x, p):
    mu = get_expectation(x, p)
    return np.sum(((x - mu)**2) * p)
```

###  標準偏差を求める関数を定義しなさい Define a function of standard deviation

```python
def get_std(x, p):
    return np.sqrt(get_variance(x, p))
```

###  期待値、分散、標準偏差を求めなさい Calculate the expected value, variance and standard deviation

```python
mean = get_expectation(x, p)
var = get_variance(x, p)
std = get_std(x, p)

print(f'{mean:.2f}')
print(f'{var:.2f}')
print(f'{std:.2f}')
```

###  標準化変数を求める関数を定義しなさい Define a function of z-score normalization

```python
def get_standardized_x(x, p):
    mu = get_expectation(x, p)
    sigma = get_std(x, p)
    return (x - mu) / sigma
```

###  標準化変数を求め、標準化後の期待値、分散、標準偏差を求めなさい Calculate the expected value, variance and standard deviation of the normalized probability distribution

```python
x_std = get_standardized_x(x, p)

mean_std = get_expectation(x_std, p)
var_std = get_variance(x_std, p)
std_std = get_std(x_std, p)

print(f'{mean_std:.2f}')
print(f'{var_std:.2f}')
print(f'{std_std:.2f}')
```

### 6. 標準化変数を求め、標準化後の確率分布を図示しなさい Plot the normalized probability distribution

```python
plt.bar(x_std, p)
plt.show()
```
